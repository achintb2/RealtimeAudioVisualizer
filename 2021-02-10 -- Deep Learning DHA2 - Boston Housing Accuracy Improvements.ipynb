```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import boston_housing
```

```python
data = boston_housing.load_data()
```

```python
(x_train, y_train), (x_test, y_test) = data
```

```python
x_test.shape
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(64, activation = 'relu', input_shape = (13, )))
model.add(keras.layers.Dense(64, activation = 'relu'))
model.add(keras.layers.Dense(1))


model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mae'])

model.fit(x_train, y_train, epochs = 100, batch_size = 10)
```

### Let's see if scaling helps with the model without changing anything else - Using Standard Scaler:

```python
from sklearn.preprocessing import StandardScaler

s = StandardScaler()
scaledTrainData = s.fit(x_train).transform(x_train)
scaledTrainLabels = s.fit(y_train.reshape(-1,1)).transform(y_train.reshape(-1,1))

scaledTestData = s.fit(x_test).transform(x_test)
scaledTestLabels = s.fit(y_test.reshape(-1, 1)).transform(y_test.reshape(-1, 1))

model = keras.Sequential()
model.add(keras.layers.Dense(64, activation = 'relu', input_shape = (13, )))
model.add(keras.layers.Dense(64, activation = 'relu'))
model.add(keras.layers.Dense(1))
model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mae'])

model.fit(scaledTrainData, scaledTrainLabels, epochs = 100, batch_size = 10)
```

### As you can see - after scaling,
### the MAE loss started from 0.5+ and ended at approx 0.09
### which is a significant improvement from the original data
### Original loss was: starting from 9 and ending at 3
### After scaling, loss is: starting from 0.4 and ending at 0.1

## Now, let's test and see if the accuracy holds up:

```python
x_predicted = model.predict(scaledTestData)
error = scaledTestLabels - x_predicted
average_error = sum(error)/len(x_predicted)
average_accuracy = 100-average_error[0]

print("With Standard Scaler:")
print(f"Average Loss is: {100-average_accuracy}")
print(f"Mean error is: {average_error[0]} %\n")
```

### The accuracy holds up well in the testing data as well.
